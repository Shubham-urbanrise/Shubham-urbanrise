# **AUGMENTED DICKEY-FULLER-TEST**

```python
import pandas as pd
from statsmodels.tsa.stattools import adfuller

# Assuming grouped_df is your DataFrame from the previous step
# Get the first campaign name from the top 20 campaigns
first_campaign = top_20_campaigns[0]

def adf_test_and_save(df, campaign_name, window_name, window_data):
    # Perform Augmented Dickey-Fuller test
    result = adfuller(window_data['Clicks'])

    # Print test statistic, p-value, and critical values
    print(f'ADF Statistic ({window_name}):', result[0])
    print(f'p-value ({window_name}):', result[1])
    print(f'Critical Values ({window_name}):')
    for key, value in result[4].items():
        print(f'   {key}: {value}')

    # Save the window data
    window_data.to_csv(f'{campaign_name}_{window_name}_data.csv', index=False)

# Filter the DataFrame for the first campaign
campaign_df = grouped_df[grouped_df['Campaign Name'] == first_campaign]

# Define the time windows
full_time_data = campaign_df
one_hundred_days_data = campaign_df[campaign_df['Date Start'] >= campaign_df['Date Start'].max() - pd.DateOffset(days=100)]
one_month_data = campaign_df[campaign_df['Date Start'] >= campaign_df['Date Start'].max() - pd.DateOffset(months=1)]

# Perform ADF test and save the data for each window
adf_test_and_save(grouped_df, first_campaign, 'Full Time', full_time_data)
adf_test_and_save(grouped_df, first_campaign, '100 Days', one_hundred_days_data)
adf_test_and_save(grouped_df, first_campaign, '1 Month', one_month_data)
```

**Explaination**

The Augmented Dickey-Fuller (ADF) test is used to determine if a unit root is present in a time series dataset, which helps in understanding if the data is stationary or not. Here's what each part of the output means:

1. **ADF Statistic:** This is the test statistic from the ADF test. It compares the actual difference in the data values to the expected difference if the data had a unit root. If the ADF statistic is less than the critical values (at 1%, 5%, and 10% levels), we can reject the null hypothesis (presence of a unit root) and conclude that the data is stationary.

2. **p-value:** This is the probability that the null hypothesis (presence of a unit root) is true. A p-value below a certain threshold (commonly 0.05) indicates that we can reject the null hypothesis and consider the data to be stationary.

3. **Critical Values:** These are the critical values at which we compare the ADF statistic. If the ADF statistic is less than these critical values, we can reject the null hypothesis.

In your case, for the "Full Time" series, the ADF statistic is -2.726, which is greater than the critical values at 1%, 5%, and 10%. The p-value is 0.070, which is also greater than 0.05. Therefore, we fail to reject the null hypothesis, suggesting that the "Full Time" series is not stationary.

For the "100 Days" and "1 Month" series, the results are similar. The ADF statistics are -2.412 and -2.428, respectively, and the p-values are 0.138 and 0.134, respectively. In both cases, we fail to reject the null hypothesis, indicating that these series are also not stationary.

In summary, based on these results, the data for all three time series ("Full Time," "100 Days," and "1 Month") are not stationary.

## **We will difference the time series and re-test on the differences:**

```python
import pandas as pd
from statsmodels.tsa.stattools import adfuller

# Assuming grouped_df is your DataFrame from the previous step
# Get the first campaign name from the top 20 campaigns
first_campaign = top_20_campaigns[0]

def adf_test_diff_and_save(df, campaign_name, window_name, window_data):
    # Perform Augmented Dickey-Fuller test on first differences
    result = adfuller(window_data['Clicks'].diff().dropna())

    # Print test statistic, p-value, and critical values
    print(f'ADF Statistic ({window_name}):', result[0])
    print(f'p-value ({window_name}):', result[1])
    print(f'Critical Values ({window_name}):')
    for key, value in result[4].items():
        print(f'   {key}: {value}')

    # Save the window data
    window_data.to_csv(f'{campaign_name}_{window_name}_data_diff.csv', index=False)

# Filter the DataFrame for the first campaign
campaign_df = grouped_df[grouped_df['Campaign Name'] == first_campaign]

# Define the time windows
full_time_data = campaign_df
one_hundred_days_data = campaign_df[campaign_df['Date Start'] >= campaign_df['Date Start'].max() - pd.DateOffset(days=100)]
one_month_data = campaign_df[campaign_df['Date Start'] >= campaign_df['Date Start'].max() - pd.DateOffset(months=1)]

# Perform ADF test on first differences and save the data for each window
adf_test_diff_and_save(grouped_df, first_campaign, 'Full Time', full_time_data)
adf_test_diff_and_save(grouped_df, first_campaign, '100 Days', one_hundred_days_data)
adf_test_diff_and_save(grouped_df, first_campaign, '1 Month', one_month_data)
```

**Explaination**
This output is the result of performing the Augmented Dickey-Fuller (ADF) test on the differenced time series data (`data.diff()`), which is a common technique to make non-stationary data stationary. Here's what each part of the output means:

1. **ADF Statistic:** This is the test statistic from the ADF test on the differenced data. A more negative ADF statistic indicates a stronger rejection of the null hypothesis (presence of a unit root), which is necessary for the data to be considered stationary.

2. **p-value:** This is the probability that the null hypothesis is true. A very small p-value (typically less than 0.05) indicates that we can reject the null hypothesis and consider the data to be stationary.

3. **Critical Values:** These are the critical values at which we compare the ADF statistic. If the ADF statistic is more negative than these critical values, we can reject the null hypothesis.

In your case, for the "Full Time" series, the ADF statistic is -7.278, which is much more negative than the critical values at 1%, 5%, and 10%. The p-value is also very small (1.528e-10), indicating a strong rejection of the null hypothesis. Therefore, we can conclude that the differenced "Full Time" series is stationary.

Similarly, for the "100 Days" and "1 Month" series, the ADF statistics are -6.937 and -5.271, respectively, and the p-values are very small as well. This suggests that both the "100 Days" and "1 Month" series, after differencing, are also stationary.

In summary, differencing the data appears to have made all three time series ("Full Time," "100 Days," and "1 Month") stationary, as indicated by the ADF test results.

# **WALD-WOLFOWITZ TEST**

```python
from statsmodels.sandbox.stats.runs import runstest_1samp

def runs_test_and_save(df, campaign_name, window_name, window_data):
    # Calculate percentage change and drop NaN values
    pct_change = window_data['Clicks'].pct_change().dropna()

    # Binarize the percentage change
    binarized = list(map(lambda x: 1.0 if x > 0.0 else 0.0, pct_change))

    # Perform runs test
    result = runstest_1samp(binarized)

    # Print test results
    print(f'Runs Test Results ({window_name}):')
    print(f'(Z-Score,p-value):{result}')
    # Save the window data
    window_data.to_csv(f'{campaign_name}_{window_name}_data.csv', index=False)

# Perform runs test and save the data for each window
runs_test_and_save(grouped_df, first_campaign, 'Full Time', full_time_data)
runs_test_and_save(grouped_df, first_campaign, '100 Days', one_hundred_days_data)
runs_test_and_save(grouped_df, first_campaign, '1 Month', one_month_data)
```

**Explaination**
The runs test is a non-parametric statistical test that checks for randomness in a dataset. It examines whether the values in the dataset are randomly distributed or if there are systematic patterns. Here's what each part of the runs test results means:

1. **Z-Score:** The Z-score measures the number of standard deviations a data point is from the mean of the dataset. In the runs test, it indicates how the runs (sequences of consecutive increasing or decreasing values) in the dataset deviate from what would be expected in a random sequence. A higher absolute Z-score suggests a greater deviation from randomness.

2. **p-value:** The p-value is the probability of obtaining the observed Z-score or more extreme values if the data were randomly distributed. A low p-value (usually less than 0.05) suggests that the data is not randomly distributed.

Now, interpreting the results:

- **Full Time Series:** The Z-score is 1.900 with a p-value of 0.057. The p-value is greater than 0.05, indicating that we fail to reject the null hypothesis that the data is random at the 5% significance level. However, it is close to 0.05, so there might be some indication of non-randomness, but it is not statistically significant.

- **100 Days Series:** The Z-score is 0.471 with a p-value of 0.637. The p-value is much greater than 0.05, indicating that we cannot reject the null hypothesis of randomness at the 5% significance level. The data appears to be random.

- **1 Month Series:** The Z-score is 0.008 with a p-value of 0.993. The p-value is very high, indicating strong evidence in favor of the null hypothesis of randomness. The data is likely to be random.

In summary, based on the runs test results, the "100 Days" and "1 Month" series appear to be random, while the "Full Time" series shows some indication of non-randomness, although not statistically significant at the 5% level.

# **ACF AND PACF PLOTS**

```python
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

def plot_acf_pacf(df, campaign_name, window_name, window_data):
    # Calculate percentage change and drop NaN values
    pct_change = window_data['Clicks'].pct_change().dropna()

    # Plot percentage change
    plt.figure(figsize=(10, 6))
    plt.plot(pct_change, label='Percentage Change')
    plt.title(f'Percentage Change - {campaign_name} ({window_name})')
    plt.legend()
    plt.show()

    # Plot ACF
    plt.figure(figsize=(10, 6))
    plot_acf(pct_change, lags=40)
    plt.title(f'Autocorrelation Function (ACF) - {campaign_name} ({window_name})')
    plt.show()

    # Plot PACF
    plt.figure(figsize=(10, 6))
    plot_pacf(pct_change, lags=40)
    plt.title(f'Partial Autocorrelation Function (PACF) - {campaign_name} ({window_name})')
    plt.show()

# Plot ACF and PACF for each window
plot_acf_pacf(grouped_df, first_campaign, 'Full Time', full_time_data)
#plot_acf_pacf(grouped_df, first_campaign, '100 Days', one_hundred_days_data)
#plot_acf_pacf(grouped_df, first_campaign, '1 Month', one_month_data)
```

```python
from statsmodels.regression.linear_model import OLS
from statsmodels.stats.stattools import durbin_watson

def durbin_watson_test(df, campaign_name, window_name, window_data):
    # Calculate differenced series and drop NaN values
    diff_series = window_data['Clicks'].diff().dropna()

    # Create the exogenous variable as a range of the same length as the differenced series
    exog = pd.Series(range(len(diff_series)), index=diff_series.index)

    # Create the OLS model
    model = OLS(diff_series, exog)
    fitted = model.fit()

    # Calculate the Durbin-Watson statistic
    dw_stat = durbin_watson(fitted.resid)

    # Print the Durbin-Watson statistic
    print(f'Durbin-Watson Statistic ({window_name}):', dw_stat)

# Perform Durbin-Watson test for each window
durbin_watson_test(grouped_df, first_campaign, 'Full Time', full_time_data)
durbin_watson_test(grouped_df, first_campaign, '100 Days', one_hundred_days_data)
durbin_watson_test(grouped_df, first_campaign, '1 Month', one_month_data)
```

**Explaination**
This is known as the Durbin-Watson statistic, and it gauges the serial correlations on a scale of 0–4. The closer to 0 the statistic, the more evidence for positive serial correlation. The closer to 4, the more evidence for negative serial correlation. A DW statistic around 2 would signal evidence for no correlations at all. Since our test returned a DW statistic of around 2.5
, we can conclude with pretty good evidence that there is no serial correlation amongst the OLS residuals and therefore the time series most likely follows a random walk.

## **LJUNG-BOX TEST**   # Giving ERROR

```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.stats.diagnostic import acorr_ljungbox

def ljung_box_test(df, campaign_name, window_name, window_data):
    try:
        # Ensure 'Clicks' column exists
        if 'Clicks' not in window_data.columns:
            print(f"Error: 'Clicks' column not found in {window_name} data for campaign {campaign_name}")
            return

        # Calculate percentage change and drop NaN values
        pct_change = window_data['Clicks'].pct_change().dropna()

        # Ensure pct_change is not empty
        if pct_change.empty:
            print(f"Error: No data available after calculating percentage change for {window_name} in campaign {campaign_name}")
            return

        # Check if there are enough data points
        if len(pct_change) < 10:  # Assuming we need at least 10 data points
            print(f"Error: Not enough data points after percentage change for {window_name} in campaign {campaign_name}")
            return

        # Reset index to avoid indexing issues
        pct_change = pct_change.reset_index(drop=True)

        # Fit ARIMA(1, 1, 1) model
        model = ARIMA(pct_change, order=(1, 1, 1))
        fitted = model.fit()

        # Perform Ljung-Box test
        lb_results = acorr_ljungbox(fitted.resid, lags=min(20, len(fitted.resid) - 1))[1]

        # Print Ljung-Box test p-values for each lag
        print(f'Ljung-Box Test p-values ({window_name}):')
        for lag, p_value in enumerate(lb_results, start=1):
            print(f'Lag {lag}: {p_value}')

    except Exception as e:
        print(f"Error fitting ARIMA model for {window_name} in campaign {campaign_name}: {e}")

# Perform Ljung-Box test for each window
ljung_box_test(grouped_df, first_campaign, 'Full Time', full_time_data)
ljung_box_test(grouped_df, first_campaign, '100 Days', one_hundred_days_data)
ljung_box_test(grouped_df, first_campaign, '1 Month', one_month_data)



```

## **BREUSCH-GODFREY TEST**

```python
import pandas as pd
from statsmodels.regression.linear_model import OLS
from statsmodels.stats.diagnostic import acorr_breusch_godfrey

def breusch_godfrey_test(df, campaign_name, window_name, window_data):
    # Calculate percentage change and drop NaN values
    pct_change = window_data['Clicks'].pct_change().dropna()

    # Ensure pct_change is not empty
    if pct_change.empty:
        print(f"Error: No data available after calculating percentage change for {window_name} in campaign {campaign_name}")
        return

    # Check if there are enough data points
    if len(pct_change) < 10:  # Assuming we need at least 10 data points
        print(f"Error: Not enough data points after percentage change for {window_name} in campaign {campaign_name}")
        return

    # Reset index to avoid indexing issues
    pct_change = pct_change.reset_index(drop=True)

    # Create the exogenous variable as a range of the same length as the pct_change series
    exog = pd.Series(range(len(pct_change)))

    # Fit OLS model
    model = OLS(pct_change, exog)
    fitted = model.fit()

    # Perform Breusch-Godfrey test
    bg_results = acorr_breusch_godfrey(fitted)

    # Print Breusch-Godfrey test results
    print(f'Breusch-Godfrey Test Results ({window_name}):')
    print(f'Lagrange multiplier statistic: {bg_results[0]}')
    print(f'p-value: {bg_results[1]}')
    print(f'f-statistic: {bg_results[2]}')
    print(f'f-pvalue: {bg_results[3]}')

# Perform Breusch-Godfrey test for each window
breusch_godfrey_test(grouped_df, first_campaign, 'Full Time', full_time_data)
breusch_godfrey_test(grouped_df, first_campaign, '100 Days', one_hundred_days_data)
breusch_godfrey_test(grouped_df, first_campaign, '1 Month', one_month_data)
```

### **Explaination**
The Breusch-Godfrey test is used to detect the presence of autocorrelation in the residuals of a regression model. Here's an explanation of each component of the Breusch-Godfrey test results:

1. **Lagrange Multiplier Statistic (LM statistic):** This statistic is used to test the null hypothesis that there is no autocorrelation up to a specified order in the residuals. A higher value of the LM statistic suggests stronger evidence against the null hypothesis.

2. **p-value:** This is the probability of obtaining the observed LM statistic (or more extreme) if the null hypothesis is true. A low p-value (typically less than 0.05) indicates that we reject the null hypothesis and conclude that there is autocorrelation in the residuals.

3. **F-Statistic:** This is another way to test for autocorrelation, calculated differently from the LM statistic. It also tests the null hypothesis that there is no autocorrelation in the residuals.

4. **F-pvalue:** The p-value associated with the F-statistic. Similar to the LM statistic p-value, a low F-pvalue indicates that we reject the null hypothesis and conclude that there is autocorrelation.

Interpreting the results:

- **Full Time Series:**
  - **LM statistic:** 4.851
  - **p-value:** 0.901
  - **F-statistic:** 0.427
  - **F-pvalue:** 0.933

  Both p-values are much higher than 0.05, indicating that we fail to reject the null hypothesis. Therefore, there is no evidence of autocorrelation in the residuals for the "Full Time" series.

- **100 Days Series:**
  - **LM statistic:** 5.008
  - **p-value:** 0.891
  - **F-statistic:** 0.216
  - **F-pvalue:** 0.994

  Similarly, both p-values are much higher than 0.05, indicating no evidence of autocorrelation in the residuals for the "100 Days" series.

- **1 Month Series:**
  - **LM statistic:** 1.893
  - **p-value:** 0.864
  - **F-statistic:** 0.248
  - **F-pvalue:** 0.935

  Both p-values are much higher than 0.05, indicating no evidence of autocorrelation in the residuals for the "1 Month" series.

In summary, the Breusch-Godfrey test results for all three time series ("Full Time," "100 Days," and "1 Month") indicate that there is no significant autocorrelation in the residuals. The high p-values for both the LM and F-statistics across all time series suggest that the residuals are not autocorrelated.

## **ONE-TAILED TEST**

```python
import numpy as np

def calculate_critical_value_and_positive_changes(window_data):
    pct_changes = window_data['Clicks'].pct_change().dropna()
    critical_value = 1.64 * (np.sqrt(len(pct_changes) / 2))
    num_positive = len([*filter(lambda x: x > 0.0, pct_changes)])
    return critical_value, num_positive

# Calculate for each window
windows = [full_time_data, one_hundred_days_data, one_month_data]

for i, window in enumerate(windows, start=1):
    critical_value, num_positive = calculate_critical_value_and_positive_changes(window)
    print(f"Window {i}: Critical Value = {critical_value}, Number of Positive Changes = {num_positive}")

```
