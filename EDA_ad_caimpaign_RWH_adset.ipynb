# **AUGMENTED DICKEY-FULLER-TEST**

```python
import pandas as pd
from statsmodels.tsa.stattools import adfuller

# Assuming grouped_df is your DataFrame from the previous step
# Get the first campaign name from the top 20 campaigns
first_adset= top_20[0]

def adf_test_and_save(df, campaign_name, window_name, window_data):
    # Perform Augmented Dickey-Fuller test
    result = adfuller(window_data['Clicks'])

    # Print test statistic, p-value, and critical values
    print(f'ADF Statistic ({window_name}):', result[0])
    print(f'p-value ({window_name}):', result[1])
    print(f'Critical Values ({window_name}):')
    for key, value in result[4].items():
        print(f'   {key}: {value}')

    # Save the window data
    window_data.to_csv(f'{campaign_name}_{window_name}_data.csv', index=False)

# Filter the DataFrame for the first campaign
adset_df = grouped_df[grouped_df['Adset Id'] == first_adset]

# Define the time windows
full_time_data = adset_df
one_hundred_days_data = adset_df[adset_df['Date Start'] >= adset_df['Date Start'].max() - pd.DateOffset(days=100)]
one_month_data = adset_df[adset_df['Date Start'] >= adset_df['Date Start'].max() - pd.DateOffset(months=1)]

# Perform ADF test and save the data for each window
adf_test_and_save(grouped_df, first_adset, 'Full Time', full_time_data)
adf_test_and_save(grouped_df, first_adset, '100 Days', one_hundred_days_data)
adf_test_and_save(grouped_df, first_adset, '1 Month', one_month_data)
```

## **Explaination**
The results you've provided are from an Augmented Dickey-Fuller (ADF) test, which is used to test for the presence of a unit root in a time series dataset. A unit root would indicate that the data is non-stationary, meaning its statistical properties like mean and variance change over time.

Here's how to interpret the results:

1. **ADF Statistic**: This is the test statistic from the ADF test. It's used to determine the significance of the results. A more negative (or less positive) value indicates stronger evidence against the null hypothesis (the presence of a unit root).

2. **p-value**: This is the probability of observing the ADF statistic if the null hypothesis were true. A p-value below a certain significance level (commonly 0.05) indicates that you can reject the null hypothesis and infer stationarity in the data. In your case, the p-values are 0.066 for the Full Time series, 0.198 for the 100 Days series, and 0.982 for the 1 Month series. These values suggest that you cannot reject the null hypothesis for the 100 Days and 1 Month series, but you can for the Full Time series at a significance level of 0.1.

3. **Critical Values**: These are the critical values at different significance levels (1%, 5%, and 10%). They are used to compare with the ADF statistic. If the ADF statistic is more extreme (i.e., further away from zero) than these critical values, you can reject the null hypothesis.

Based on these results:
- For the Full Time series, you can reject the null hypothesis at a significance level of 0.1, suggesting that the series is stationary.
- For the 100 Days and 1 Month series, you fail to reject the null hypothesis, indicating that these series may have a unit root and are non-stationary.

# **WALD-WOLFOWITZ TEST**

```python
from statsmodels.sandbox.stats.runs import runstest_1samp

def runs_test_and_save(df, campaign_name, window_name, window_data):
    # Calculate percentage change and drop NaN values
    pct_change = window_data['Clicks'].pct_change().dropna()

    # Binarize the percentage change
    binarized = list(map(lambda x: 1.0 if x > 0.0 else 0.0, pct_change))

    # Perform runs test
    result = runstest_1samp(binarized)

    # Print test results
    print(f'Runs Test Results ({window_name}):')
    print(f'(Z-Score,p-value):{result}')
    # Save the window data
    window_data.to_csv(f'{campaign_name}_{window_name}_data.csv', index=False)

# Perform runs test and save the data for each window
runs_test_and_save(grouped_df, first_adset, 'Full Time', full_time_data)
runs_test_and_save(grouped_df, first_adset, '100 Days', one_hundred_days_data)
runs_test_and_save(grouped_df, first_adset, '1 Month', one_month_data)
```

## **Explaination**
The Runs Test is a non-parametric statistical test that checks for randomness in a sequence of data. It examines whether the values tend to alternate rather than clustering together. Here's how to interpret the results:

1. **Z-Score**: The Z-Score measures how many standard deviations an element is from the mean of the distribution. For the Runs Test, a higher absolute Z-Score indicates a stronger deviation from randomness.

2. **p-value**: The p-value associated with the Z-Score indicates the probability of obtaining a Z-Score as extreme as the one observed, assuming that the sequence is random. A low p-value suggests that the sequence is not random.

Here's an interpretation of the results for each time frame:

- **Full Time**: The Z-Score is 3.61, which is quite high, indicating a significant deviation from randomness. The p-value is very low (0.00031), suggesting that the sequence is not random.

- **100 Days**: The Z-Score is 0.36, which is close to zero, indicating little deviation from randomness. The p-value is 0.72, which is high, suggesting that the sequence is likely random.

- **1 Month**: The Z-Score is 1.65, indicating a moderate deviation from randomness. The p-value is 0.0999, which is close to 0.1. Generally, a p-value above 0.1 is considered not statistically significant, so there is some evidence that the sequence is random, but it's not strong.

In summary, the Runs Test results suggest that the Full Time series is significantly non-random, while the 100 Days series is likely random, and the 1 Month series shows some deviation from randomness but is not statistically significant.

# **DURBIN-WATSON TEST**

```python
from statsmodels.regression.linear_model import OLS
from statsmodels.stats.stattools import durbin_watson

def durbin_watson_test(df, campaign_name, window_name, window_data):
    # Calculate differenced series and drop NaN values
    diff_series = window_data['Clicks'].diff().dropna()

    # Create the exogenous variable as a range of the same length as the differenced series
    exog = pd.Series(range(len(diff_series)), index=diff_series.index)

    # Create the OLS model
    model = OLS(diff_series, exog)
    fitted = model.fit()

    # Calculate the Durbin-Watson statistic
    dw_stat = durbin_watson(fitted.resid)

    # Print the Durbin-Watson statistic
    print(f'Durbin-Watson Statistic ({window_name}):', dw_stat)

# Perform Durbin-Watson test for each window
durbin_watson_test(grouped_df, first_adset, 'Full Time', full_time_data)
durbin_watson_test(grouped_df, first_adset, '100 Days', one_hundred_days_data)
durbin_watson_test(grouped_df, first_adset, '1 Month', one_month_data)
```

**Explaination**
The Durbin-Watson statistic is a test for autocorrelation in the residuals from a regression analysis. It ranges in value from 0 to 4. A value of 2 indicates no autocorrelation, while values significantly below 2 indicate positive autocorrelation, and values significantly above 2 indicate negative autocorrelation.

Here's how to interpret the results:

- **Full Time**: The Durbin-Watson statistic is 2.199, which is close to 2, indicating little autocorrelation in the residuals.

- **100 Days**: The Durbin-Watson statistic is 2.091, also indicating little autocorrelation in the residuals for the 100 Days series.

- **1 Month**: The Durbin-Watson statistic is 2.659, which is higher than 2, indicating the possibility of negative autocorrelation in the residuals for the 1 Month series.

The significance of these values depends on the context of your analysis and the specific assumptions of your model. In general, values close to 2 suggest that the residuals are independent, which is desirable in regression analysis. However, if you are working with time series data, autocorrelation in the residuals may indicate that your model is not capturing all the relevant information in the data.

## **BREUSCH-GODFREY TEST**

```python
import pandas as pd
from statsmodels.regression.linear_model import OLS
from statsmodels.stats.diagnostic import acorr_breusch_godfrey

def breusch_godfrey_test(df, campaign_name, window_name, window_data):
    # Calculate percentage change and drop NaN values
    pct_change = window_data['Clicks'].pct_change().dropna()

    # Ensure pct_change is not empty
    if pct_change.empty:
        print(f"Error: No data available after calculating percentage change for {window_name} in campaign {campaign_name}")
        return

    # Check if there are enough data points
    if len(pct_change) < 10:  # Assuming we need at least 10 data points
        print(f"Error: Not enough data points after percentage change for {window_name} in campaign {campaign_name}")
        return

    # Reset index to avoid indexing issues
    pct_change = pct_change.reset_index(drop=True)

    # Create the exogenous variable as a range of the same length as the pct_change series
    exog = pd.Series(range(len(pct_change)))

    # Fit OLS model
    model = OLS(pct_change, exog)
    fitted = model.fit()

    # Perform Breusch-Godfrey test
    bg_results = acorr_breusch_godfrey(fitted)

    # Print Breusch-Godfrey test results
    print(f'Breusch-Godfrey Test Results ({window_name}):')
    print(f'Lagrange multiplier statistic: {bg_results[0]}')
    print(f'p-value: {bg_results[1]}')
    print(f'f-statistic: {bg_results[2]}')
    print(f'f-pvalue: {bg_results[3]}')

# Perform Breusch-Godfrey test for each window
breusch_godfrey_test(grouped_df, first_adset, 'Full Time', full_time_data)
breusch_godfrey_test(grouped_df, first_adset, '100 Days', one_hundred_days_data)
breusch_godfrey_test(grouped_df, first_adset, '1 Month', one_month_data)
```

**Explaination**
The Breusch-Godfrey test is used to test for autocorrelation in the residuals of a regression model, especially in time series data. It extends the Durbin-Watson test by allowing for higher-order autocorrelation. Here's how to interpret the results:

1. **Lagrange Multiplier Statistic**: This statistic measures the overall significance of the autocorrelation in the residuals. Higher values indicate stronger evidence of autocorrelation.

2. **p-value**: The p-value associated with the Lagrange multiplier statistic indicates the probability of observing the statistic if the null hypothesis of no autocorrelation is true. A low p-value suggests that you can reject the null hypothesis.

3. **F-Statistic**: The F-statistic tests the joint hypothesis that all autocorrelation coefficients up to a certain lag are zero. It complements the Lagrange multiplier statistic.

4. **F-Pvalue**: This is the p-value associated with the F-statistic. As with the Lagrange multiplier p-value, a low F-pvalue indicates evidence against the null hypothesis.

Now, interpreting the results for each time frame:

- **Full Time**: The Lagrange multiplier statistic is 6.06 with a p-value of 0.81. This suggests that there is no significant autocorrelation in the residuals for the Full Time series.

- **100 Days**: The Lagrange multiplier statistic is 4.99 with a p-value of 0.89. Similar to the Full Time series, this indicates no significant autocorrelation in the residuals for the 100 Days series.

- **1 Month**: The Lagrange multiplier statistic is 5.64 with a p-value of 0.34. Again, this suggests no significant autocorrelation in the residuals for the 1 Month series.

In summary, based on the Breusch-Godfrey test results, there is no significant autocorrelation in the residuals for any of the three time frames, indicating that the regression models used for these series adequately capture the autocorrelation patterns.

## **ONE-TAILED TEST**

```python
import numpy as np

def calculate_critical_value_and_positive_changes(window_data):
    pct_changes = window_data['Clicks'].pct_change().dropna()
    critical_value = 1.64 * (np.sqrt(len(pct_changes) / 2))
    num_positive = len([*filter(lambda x: x > 0.0, pct_changes)])
    return critical_value, num_positive

# Calculate for each window
windows = [full_time_data, one_hundred_days_data, one_month_data]

for i, window in enumerate(windows, start=1):
    critical_value, num_positive = calculate_critical_value_and_positive_changes(window)
    print(f"Window {i}: Critical Value = {critical_value}, Number of Positive Changes = {num_positive}")

```
